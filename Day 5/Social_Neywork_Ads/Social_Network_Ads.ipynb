```python
X= df.drop(['User ID','Purchased'],axis=1)
y= df['Purchased']
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X , y , test_size=0.3 , random_state=42)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
dic={
    'LogisticRegression':LogisticRegression(),
    'RandomeForestClassifier':RandomForestClassifier(max_depth=5, n_estimators=7, random_state=0),
    'KNeighborsClassifier':KNeighborsClassifier(n_neighbors=15,metric='manhattan'),
    'DecisionTreeClassifier':DecisionTreeClassifier(criterion='entropy', max_depth=2),
    'Support Vector Machine':SVC(kernel='linear'),
}

from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
def accuracy(y_test,y_pred):
    acc = accuracy_score(y_test,y_pred)
    return acc
    
accuracy_models=[]

models_names = []
for model_name , model in dic.items():
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    accuracy_models.append(accuracy(y_test,y_pred))
    models_names.append(model_name)
    print(classification_report(y_test,y_pred))
    print(confusion_matrix(y_test,y_pred))
    print('++++++++++++++++++++++++++++++++++++++')
```

```python
accuracy_df = pd.DataFrame()
accuracy_df['models'] =models_names
accuracy_df['Accuray'] =accuracy_models
accuracy_df = accuracy_df.set_index('models')
accuracy_df
```
