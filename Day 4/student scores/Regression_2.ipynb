```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
file=pd.read_csv("/content/student_scores (1).csv")
```

```python
file=pd.read_csv("/content/student_scores (1).csv")
file=file.fillna(np.mean(file["Hours"]))
```

```python
file
```

```python
from sklearn.linear_model import LinearRegression

x=file[["Hours"]]
y=file["Scores"]
model=LinearRegression()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.09,random_state=42)
model.fit(x_train,y_train)
yp=model.predict(x_test)
model.score(x_test,y_test)
```

```python
xplot=plt.scatter(x_train,y_train,label="h",color="red")
yplot=plt.plot(x_test,yp,label="S",color="blue")
plt.legend()
plt.show()
```
