```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path=r'student_scores.csv'
data = pd.read_csv(file_path)
print(data.head())
data=data.fillna(np.mean(data['Hours']))
```

```python
x=data[['Hours']]
y=data['Scores']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
plt.scatter(x,y, color='red',label='Data points')
plt.plot(x_test, y_pred, color='blue', label='Regression line')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Linear Regression : Hours vs Scores')
plt.legend()
plt.show()





```

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
```

```python

# Polynomial regression
degree = 3
poly = PolynomialFeatures(degree=degree)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

model = LinearRegression()
model.fit(x_train_poly, y_train)

# Predict for test set and entire dataset
y_pred = model.predict(x_test_poly)
x_poly = poly.transform(x)
y_poly_pred = model.predict(x_poly)

# Sort x and corresponding predictions for smooth plotting
sorted_indices = np.argsort(x.values.ravel())
x_sorted = x.values.ravel()[sorted_indices]
y_sorted = y_poly_pred[sorted_indices]

# Plot original data points
plt.scatter(x, y, color='red', label='Data points')

# Plot polynomial regression line as a single continuous curve
plt.plot(x_sorted, y_sorted, color='blue', label=f'Polynomial Regression (degree={degree})')

# Customize plot
plt.xlabel('Hours Studied')
plt.ylabel('Scores Achieved')
plt.title('Polynomial Regression: Hours vs Scores')
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```
