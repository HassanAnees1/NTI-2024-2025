```python
# Important libraries
from sklearn.datasets import load_iris # images iris
from sklearn import tree # model decision tree

x , y = load_iris(return_X_y=True) # return data split x , y
# can get all data and split manuall
print('All Features : \n' , x)
print(' Target : \n', y) # 3 categories
```

```python
# tree
clf = tree.DecisionTreeClassifier(criterion='entropy' , max_depth=100) # call decision tree
clf = clf.fit(x,y) # fit model

tree.plot_tree(clf , filled=True ) # show tree
# in show tree use gini not entropy can change function to entropy
```

```python
# predect on this folwer [4.5 2.2 1.7 0.2]

import numpy as np
from sklearn.datasets import load_iris
from sklearn import tree

# Input the flower features as a NumPy array.
flower_features = np.array([[4.5, 2.2, 1.7, 0.2]])

# Predict the flower species.
prediction = clf.predict(flower_features)

# Print the predicted species.
print("Prediction:", prediction)

iris = load_iris() # reload iris with target_names
print("Predicted Species:", iris.target_names[prediction[0]])
```
