```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.martrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/mobile_price_train.csv')

df.head()
```

```python
correlation_matrix = df.corr()
price_correlation = correlation_matrix['price_range'].sort_values(ascending=False)
price_correlation
```

```python
features = ['ram', 'battery_power', 'px_width', 'px_height']

target = ['price_range']

x=df[features]
y=df[target]

clf = tree.DecisionTreeClassifier(criterion="entropy" )
clf = clf.fit(x,y)
tree.plot_tree(clf, filled=True)


accuracy_score(y, clf.predict(x))



new_data = pd.DataFrame({'ram': [2048], 'battery_power': [4000], 'px_width': [1080], 'px_height': [1920]})


```

```python
prediction = clf.predict(new_data)
print(prediction)
```
