```python
# prompt: write a code to calculate the dependcy of price_range on all the features  and show which one is more dependable

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score

file_path = '/content/mobile_price_train.csv'
data = pd.read_csv(file_path)

# Separate features (X) and target variable (y)
X = data.drop('price_range', axis=1)
y = data['price_range']

# Calculate dependencies using Linear Regression
dependencies = {}
for col in X.columns:
    model = LinearRegression()
    model.fit(X[[col]], y)  # Train on a single feature
    r2 = r2_score(y, model.predict(X[[col]]))
    dependencies[col] = r2

# Find the feature with the highest dependency (R-squared value)
most_dependent_feature = max(dependencies, key=dependencies.get)
highest_dependency = dependencies[most_dependent_feature]

print("Feature Dependencies:")
for feature, dependency in dependencies.items():
    print(f"{feature}: {dependency}")

print(f"\nMost Dependent Feature: {most_dependent_feature} (R-squared: {highest_dependency})")
```

```python
# prompt: i want to find a correlation related to price _range and arrange them decesinding

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score

file_path = '/content/mobile_price_train.csv'
data = pd.read_csv(file_path)

# Separate features (X) and target variable (y)
X = data.drop('price_range', axis=1)
y = data['price_range']

# Calculate dependencies using Linear Regression
dependencies = {}
for col in X.columns:
    model = LinearRegression()
    model.fit(X[[col]], y)  # Train on a single feature
    r2 = r2_score(y, model.predict(X[[col]]))
    dependencies[col] = r2

# Sort dependencies in descending order
sorted_dependencies = dict(sorted(dependencies.items(), key=lambda item: item[1], reverse=True))

print("Feature Dependencies (Descending Order):")
for feature, dependency in sorted_dependencies.items():
    print(f"{feature}: {dependency}")
```

```python
# prompt: assign ram battery_power px width px height to x and the target to price range

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score

file_path = '/content/mobile_price_train.csv'
data = pd.read_csv(file_path)

# Select specific features and target variable
x = data[['ram', 'battery_power', 'px_width', 'px_height']]
y = data['price_range']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# prompt: Model logistic regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initialize and train the Logistic Regression model
model = LogisticRegression(max_iter=1000) # Increased max_iter for convergence
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)

# Evaluate the model's accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Logistic Regression: {accuracy}")
```

```python
new_data = pd.DataFrame({'ram': [2], 'battery_power': [20], 'px_width': [1080], 'px_height': [5000]})
predicted_price_range = model.predict(new_data)
print(f"Predicted price range for new data: {predicted_price_range[0]}")
```
